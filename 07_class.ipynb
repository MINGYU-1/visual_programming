```python
class SharedFund:
    BALANCE = 0
    
    def __init__(self):
        return "현재 금액: " + str(self.my_balance)
    def __str__(self):
        return "현재 금액: "+str(self.my_balance)
    def save(self,amount):
        SharedFund.BALANCE +=amount
        self.my_balance -= amount
    def withdrawl(self, amount):
        SharedFund.BALANCE -= amount
        self.my_balance += amount
```

```python
class Complex:
    def __init__(self,r,i):
        self.r = r
        self.i = i
    def __repr__(self):
        return str(self.r) + '+' + str(self.i) + 'i'
    def __add__(self,value):
        return Complex(r = self.r + value.r, 
                       i = self.i + value.i)

    
a = Complex(1,-1)
b = Complex(13,-10)
print(a.__add__(b))
```
